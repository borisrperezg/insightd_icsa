```python
#df = pd.read_csv('/gdrive/My Drive/InsighTD-AnalisisColombia/EASE_Data4RQ3_4Jupyter.csv', sep="|", index_col=False)
df = pd.read_csv('Data_RQ4.csv', sep="|", index_col=False)
df.columns.tolist()
```

```python
df
```

```python
df['TDTYPE'].replace(['TEST', 'TESTE'], 'TEST DEBT', inplace=True)
df['TDTYPE'].replace(['DESIGN', 'PROJETO'], 'DESIGN DEBT', inplace=True)
df['TDTYPE'].replace(['ARQUITETURA', 'ARCHITECTURE', 'ARCHICTETURE'], 'ARCHITECTURE DEBT', inplace=True)
df['TDTYPE'].replace(['CODE', 'CÓDIGO'], 'CODE DEBT', inplace=True)
df['TDTYPE'].replace(['DOCUMENTATION', 'DOCUMENTAÇÃO'], 'DOCUMENTATION DEBT', inplace=True)
df['TDTYPE'].replace(['REQUIREMENT', 'REQUIREMENTS', 'REQUISITOS', 'REQUISITO', 'REQUIREMENTS DEBT'], 'REQUIREMENT DEBT', inplace=True)
df['TDTYPE'].replace(['INFRASTRUCTURE', 'INFRAESTRUTURA'], 'INFRASTRUCTURE DEBT', inplace=True)
df['TDTYPE'].replace(['BUILD'], 'BUILD DEBT', inplace=True)
df['TDTYPE'].replace(['SERVIDE DEBT', 'SERVIÇO'], 'SERVICE DEBT', inplace=True)
df['TDTYPE'].replace(['USABILIDADE', 'USABILITY'], 'USABILITY DEBT', inplace=True)
df['TDTYPE'].replace(['DEFEITO', 'DEFECT'], 'DEFECT DEBT', inplace=True)
df['TDTYPE'].replace(['PROCESS', 'PROCESSO'], 'PROCESS DEBT', inplace=True)
df['TDTYPE'].replace(['AUTOMAÇÃO DE TESTES'], 'AUTOMATION TEST DEBT', inplace=True)
df['TDTYPE'].replace(['VERSIONAMENTO'], 'VERSIONING DEBT', inplace=True)
```
